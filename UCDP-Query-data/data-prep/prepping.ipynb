```python
main_df = main_df[main_df['GWNOLoc'].notna()]
```

```python
def map_locations_to_coordinates(location_str):
    if not location_str or pd.isna(location_str):  # Handle empty or NaN values
        return []
    codes = [code.strip() for code in location_str.split(',')]
    return [code_to_coords[code] for code in codes if code in code_to_coords]
```

```python
# Prepare GeoJSON features
features = []
for _, row in main_df.iterrows():
    locations = map_locations_to_coordinates(row['GWNOLoc'])
    geometry = {
        "type": "MultiPoint" if len(locations) > 1 else "Point",
        "coordinates": locations if len(locations) > 1 else locations[0]
    }
    feature = {
        "type": "Feature",
        "properties": {
            "name": row['name'],
            # Add other properties from the row as needed
        },
        "geometry": geometry
    }
    features.append(feature)
```

```python
# Example structure for `country_df`:
# code,coordinates
# US,"[-98.5833, 39.8333]"
# DE,"[10.4515, 51.1657]"

# Ensure coordinates are parsed correctly (as lists)
country_df['coordinates'] = country_df['coordinates'].apply(json.loads)

# Create a dictionary for fast mapping of codes to coordinates
code_to_coords = country_df.set_index('code')['coordinates'].to_dict()

# Function to process gwlocation and map to coordinates
def map_locations_to_coordinates(location_str):
    codes = [code.strip() for code in location_str.split(',')]
    return [code_to_coords[code] for code in codes if code in code_to_coords]

# Prepare GeoJSON features
features = []
for _, row in main_df.iterrows():
    locations = map_locations_to_coordinates(row['gwlocation'])
    geometry = {
        "type": "MultiPoint" if len(locations) > 1 else "Point",
        "coordinates": locations if len(locations) > 1 else locations[0]
    }
    feature = {
        "type": "Feature",
        "properties": {
            "name": row['name'],
            # Add other properties from the row as needed
        },
        "geometry": geometry
    }
    features.append(feature)

# Create the GeoJSON FeatureCollection
geojson = {
    "type": "FeatureCollection",
    "features": features
}

# Save GeoJSON to a file
with open("output.geojson", "w") as f:
    json.dump(geojson, f)

print("GeoJSON file created: output.geojson")
```
